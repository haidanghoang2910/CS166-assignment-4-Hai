```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
import scipy.stats as sts
```

### Problem 1: Buffon's needle 1

```python
#let d=10
#since we are having d = 2L for the approximation: L=5
d = 10
l = 5
```

```python
count=0
error = []
for i in range(1000000): #1000000 tries
    x = random.random()*5 #x runs from 0 to 5 because it has to be equal or lower than d/2
    theta = random.random()*math.pi/2 #the angle will always be equal to or smaller than 90 degrees (or pi/2)
    if x < math.sin(theta)*l/2: #the given expression in the prompt
        count +=1
    else:
        pass
    error.append(count/(i+1)-(1/math.pi)) #error compared to the actual 1/pi
estimation = count/1000000 #our estimation of 1/pi
print("Our estimation of 1/pi is", estimation)
```

```python
x_axes = np.linspace(0,1000001, 1000000)
log_x_axes = np.log(x_axes)
log_error = np.log(error)
plt.plot(log_x_axes,log_error)
plt.title("Error over number of trials")
plt.xlabel("Number of trials")
plt.ylabel("Error")
```

### Problem 3: Order Statistics

```python
#for k=1, n=9
lst_sample=[]
for i in range(1000): 
    lst=[]
    for j in range(9):
        lst.append(np.random.uniform(0,1))
    lst_sample.append(min(lst))
```

```python
plt.figure()
plt.hist(lst_sample, density=True)
x_axis = np.linspace(0,1,1000)
plt.plot(x_axis,sts.beta.pdf(x_axis, 1, 9))
plt.xlabel('value')
plt.ylabel('density')
plt.show()
```

We see that it matches quite closely with the corresponding beta distribution.

### Problem 4: Markov chains

We see that it matches quite closely with the corresponding beta distribution.

```python
#the given Markov matrix
p = np.array([[1-(math.e**-3)/2,(math.e**-3)/2,0,0], [0.5,0,0.5,0], [0,(math.e**-1)/2,1-0.5-(math.e**-1)/2,0.5], 
              [0,0,(math.e**-1)/2,1-(math.e**-1)/2]])
p
```

```python
#The eigenvector using Python: 
eig = np.linalg.eig(p.transpose())
eig
```

```python
print(eig[1][:,2]/sum(eig[1][:,2])) #normalizing the vector to represent probabilities
```

```python
#finding the eigenvector manually: we pick a random starting point, then multiply with p
#until we achieve the invariant distribution
start=[1,0,0,0]
lst_markov=[]
lst_markov.append(np.matmul(start,p))
for i in range(1000):
    current_state=lst_markov[i]
    next_state=np.matmul(current_state,p) #obtain next state by matrix-multiplying matrix with current state
    lst_markov.append(next_state)
lst_markov[-1] #take the last element - this should be closest to the invariant distribution
```

```python
lst_markov
```

We see that the distribution that we found is <0.64, 0.03, 0.09, 0.24> - this is in line with what we found with the Python function above. 

```python
#Histogram: this is a discrete distribution. We randomly take 4 values: 1,2,3,4 corresponding to four probabilities
lst=[]
dist=lst_markov[-1]
dist_cumsum = np.cumsum(dist)
for i in range(10000):
    p = random.random()
    if p <= dist_cumsum[0]:
        lst.append(1)
    elif dist_cumsum[0] < p <= dist_cumsum[1]:
        lst.append(2)
    elif dist_cumsum[1] < p <= dist_cumsum[2]:
        lst.append(3)
    else:
        lst.append(4)
```

```python
plt.hist(lst, bins=4)
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```
